# Word segmentation

```python
# 필요한 library import
import re
import codecs
```

```python
# 미리 준비한 뉴스 글 모음 파일 열기
news1 = codecs.open("ex1.txt", "r", "utf-8")
contents1 = news1.read()
```

```python
contents1
```

```python
rex = '[\[\]\(\)\(\)’‘·,…”“:\-\'+|=#★▲ⓒ>■◆&</△""©!...]'
email1 = '[a-zA-Z]+@[a-zA-Z]+\.[a-zA-Z]+'
email2 = '[a-zA-Z]+@[a-zA-Z]+'
email3 = '[a-zA-Z]+@'
site1 = '[a-zA-Z]+\.[a-zA-Z]+'
site2 = '\.[a-zA-Z]+'
```

```python
# 특수문자 및 개행문자 삭제
parse1 = re.sub(email1, ' ', contents1)
parse1 = re.sub(email2, ' ', parse1)
parse1 = re.sub(site1, ' ', parse1)
parse1 = re.sub(site2, ' ', parse1)
parse1 = re.sub(rex, ' ', parse1)
parse1 = parse1.replace("\n", "")
parse1 = parse1.replace("▶","")
parse1 = parse1.replace("\\", "")
parse1 = parse1.replace("\r", "")
parse1 = parse1.replace("\t", "")
parse1 = parse1.replace("\ufeff", "")
parse1
```

```python
parse1_word = parse1.split(" ") # 공백을 기준으로 단어 분리
for i in range(len(parse1_word)):    
    parse1_word[i] = re.sub("으로$", '', parse1_word[i])
    parse1_word[i] = re.sub("까지$", '', parse1_word[i])
    parse1_word[i] = re.sub("에서는$", '', parse1_word[i])
    parse1_word[i] = re.sub("에는$", '', parse1_word[i])
    parse1_word[i] = re.sub("에$", '', parse1_word[i])
    parse1_word[i] = re.sub("와$", '', parse1_word[i])
    parse1_word[i] = re.sub("과$", '', parse1_word[i])
    parse1_word[i] = re.sub("을$", '', parse1_word[i])
    parse1_word[i] = re.sub("를$", '', parse1_word[i])
    parse1_word[i] = re.sub("의$", '', parse1_word[i])
    
parse1_word = [parse1_word[i] for i in range(len(parse1_word)) if len(parse1_word[i]) > 1] # 한 글자 이상인 단어는 제거
# print(type(parse1_word))
parse1_str = " ".join(parse1_word)
parse1_str
# print(type(parse1_str))

# FILE_NAME = "parse1_postposition_remove.txt"
# with open(FILE_NAME, "w", encoding="utf8") as f:
#     f.write(parse1_str)
# parse1_word
```

```python
# 단어 공백으로 분리하기
# print(type(parse1_word[1]))
parse1_word.sort()
FILE_NAME = "parse1_comp_count.txt"
# with open(FILE_NAME, "w", encoding="utf8") as f:
#     for i in range(len(parse1_word)):
#         f.write(parse1_word[i]+"\n")

print("The Number of token : ",len(parse1_word))
parse1_word
```

```python
# 단어 빈도수 만들어보기
table1 = dict()
for word in parse1_word:
#     print(word)
    if word not in table1:
        table1[word] = 1
    else:
        table1[word] += 1

print("The number of type : ", len(table1))

for i,j in table1.items():
    print(i,"\t",j)
        
DICT_NAME = "ex1_comp_dict.txt"

# with open(DICT_NAME, "w", encoding="utf8") as f:
#     for i, j in table1.items():
#         f.write(i+ " : "+ str(j)+" // ")
```

```python
# 직접 분리한 글 토큰 및 타입 수
news1_comp = codecs.open("ex1_hand.txt", "r", "utf-8")
contents1_comp = news1_comp.read()
```

```python
# 최소한의 특수문자 제거
contents1_comp = contents1_comp.replace("\ufeff", " ")
contents1_comp = contents1_comp.replace("\r", " ")
contents1_comp = contents1_comp.replace("\n", " ")
contents1_comp
```

```python
# 단어 리스트 만들기
contents1_comp_list = contents1_comp.split(" ")
contents1_comp_word = [contents1_comp_list[i] for i in range(len(contents1_comp_list)) if len(contents1_comp_list[i]) > 1] # 한 글자 이상인 단어는 제거
print("Token : ", len(contents1_comp_word))
contents1_comp_word
```

```python
# 단어 빈도수 만들어보기
table1_comp = dict()
for word in contents1_comp_word:
#     print(word)
    if word not in table1_comp:
        table1_comp[word] = 1
    else:
        table1_comp[word] += 1

print("The number of type : ", len(table1_comp))

for i,j in table1_comp.items():
    print(i,"\t",j)
```

```python
news = codecs.open("news.txt", "r", "utf-8")
contents = news.read()
contents
```

```python
parse = re.sub(email1, ' ', contents)
parse = re.sub(email2, ' ', parse)
parse = re.sub(site1, ' ', parse)
parse = re.sub(site2, ' ', parse)
parse = re.sub(rex, ' ', parse)
parse = parse.replace("\n", "")
parse = parse.replace("▶","")
parse = parse.replace("\\", "")
parse = parse.replace("\r", "")
parse = parse.replace("\t", "")
parse
```

```python
parse_word = parse.split(" ") # 공백을 기준으로 단어 분리
parse_word = [parse_word[i] for i in range(len(parse_word)) if len(parse_word[i]) > 1] # 한 글자 이상인 단어는 제거
parse_word.sort()
FILE_NAME = "words_count.txt"
with open(FILE_NAME, "w", encoding="utf8") as f:
    for i in range(len(parse_word)):
        f.write(parse_word[i]+"\n")
print("Token의 수 : ", len(parse_word))
parse_word
```

```python
# 단어 빈도수 만들어보기
table = dict()
for word in parse_word:
#     print(word)
    if word not in table:
        table[word] = 1
    else:
        table[word] += 1
        

DICT_NAME = "words_dict.txt"

with open(DICT_NAME, "w", encoding="utf8") as f:
    for i, j in table.items():
        f.write(i+ "\t"+ str(j)+"\n")
        
print("Type의 수 : ", len(table))        
for i,j in table.items():
    print(i,"\t",j)
```
