# Word segmentation

```python
import re
import codecs
```

```python
news1 = codecs.open("ex1.txt", "r", "utf-8")
contents1 = news1.read()
```

```python
contents1
```

```python
rex = '[\[\]\(\)\(\)’‘·,…”“:\-\'+|=#★▲ⓒ>■◆&</△]'
email1 = '[a-zA-Z]+@[a-zA-Z]+\.[a-zA-Z]+'
email2 = '[a-zA-Z]+@[a-zA-Z]+'
site1 = '[a-zA-Z]+\.[a-zA-Z]+'
site2 = '\.[a-zA-Z]+'
```

```python
parse1 = re.sub(rex, ' ', contents1)
parse1
```

```python
news2 = codecs.open("ex2.txt", "r","utf-8")
contents2 = news2.read()
contents2
```

```python
parse2 = re.sub(rex, ' ', contents2)
# try:
#     parse2 = parse2[:parse2.index('▶')]
# except:
#     pass

parse2 = parse2.replace("\n", "")
parse2 = parse2.replace("▶", "")
parse2 = parse2.replace("\\", "")
parse2
```

```python
news = codecs.open("news.txt", "r", "utf-8")
contents = news.read()
contents
```

```python
parse = re.sub(email1, ' ', contents)
parse = re.sub(email2, ' ', parse)
parse = re.sub(site1, ' ', parse)
parse = re.sub(site2, ' ', parse)
parse = re.sub(rex, ' ', parse)
parse = parse.replace("\n", "")
parse = parse.replace("▶","")
parse = parse.replace("\\", "")
parse = parse.replace("\r", "")
parse
```

```python
parse_word = parse.split(" ")
parse_word
```
